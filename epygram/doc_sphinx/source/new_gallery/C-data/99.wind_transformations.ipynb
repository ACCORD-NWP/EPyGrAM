```python
# for figures in notebook
%matplotlib inline

# import & initialize epygram
import epygram
epygram.init_env()
```

```python
import os
INPUTS_DIR = os.path.join(epygram.config.installdir, 'doc_sphinx', 'source', 'new_gallery', 'inputs')
# open resource in read mode
r = epygram.formats.resource(os.path.join(INPUTS_DIR, 'rst.atm.P18500101.tl127.AGCM.fa'), 'r')
```

```python
psi = r.readfield('S070FONC.COURANT')
khi = r.readfield('S070POT.VITESSE')
uv = epygram.fields.psikhi2uv(psi, khi)
```

```python
fig, ax = uv.cartoplot(subsampling=5, vector_plot_method='quiver', components_are_projected_on='grid')
```
