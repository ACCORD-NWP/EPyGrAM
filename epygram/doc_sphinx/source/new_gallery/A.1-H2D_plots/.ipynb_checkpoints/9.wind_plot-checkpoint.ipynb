# Plotting wind: barbs, wind speed

```python
# for figures in notebook
%matplotlib inline

# import & initialize epygram
import epygram
epygram.init_env()
```

```python
import os
INPUTS_DIR = os.path.join(epygram.config.installdir, 'doc_sphinx', 'source', 'new_gallery', 'inputs')
# open resource in read mode
r = epygram.formats.resource(os.path.join(INPUTS_DIR, 'rst.atm.P18500101.tl127.AGCM.fa'), 'r')
```

```python
u = r.readfield('CLSVENT.MERIDIEN')
v = r.readfield('CLSVENT.ZONAL')
uv = epygram.fields.make_vector_field(u,v)
```

```python
fig, ax = uv.cartoplot(subsampling=10, vector_plot_method='barbs', subzone='CI')
```

```python
# or if you don't want wind speed:
fig, ax = uv.cartoplot(subsampling=10, vector_plot_method='barbs', subzone='CI', plot_method=None)
```

```python
# or if you want to plot module only:
ff = uv.to_module()
fig, ax = ff.cartoplot(subzone='CI')
```
