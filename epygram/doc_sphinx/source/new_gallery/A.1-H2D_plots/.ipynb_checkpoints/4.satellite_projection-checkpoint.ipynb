```python
# for figures in notebook
%matplotlib inline

# import & initialize epygram
import epygram
epygram.init_env()
```

```python
import os
INPUTS_DIR = os.path.join(epygram.config.installdir, 'doc_sphinx', 'source', 'new_gallery', 'inputs')
# open resource in read mode
r = epygram.formats.resource(os.path.join(INPUTS_DIR, 'ICMSHAROM+0022'), 'r')
```

```python
f = r.readfield('SURFTEMPERATURE')
```

```python
from cartopy.crs import NearsidePerspective
proj = NearsidePerspective(central_longitude=0.,
                           central_latitude=45.,
                           satellite_height=1e6)
fig, ax = f.cartoplot(projection=proj, extent='global')
```
