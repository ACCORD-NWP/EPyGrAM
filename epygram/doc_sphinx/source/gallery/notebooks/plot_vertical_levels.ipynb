```python
%matplotlib inline
import matplotlib.pyplot as plt
import epygram
epygram.init_env()
```

```python
r = epygram.formats.resource('/home/mary/worktmp/historic.arome.guyane-02km50+0006:00.fa', 'r')
```

```python
p = r.extractprofile('S*TEMPERATURE', -52, 5.5, vertical_coordinate=103)
```

```python
g = r.readfield('SPECSURFGEOPOTEN')
g.sp2gp()
print(g.getvalue_ll(-52,5.5)/9.81)
alt = p.geometry.vcoordinate.levels
```

```python
levels = range(1, len(alt)+1)
print(alt)
```

```python
fig, ax = plt.subplots(1,1, figsize=(16,12))
min_alt = -200
max_alt = 20000
min_level = -1
max_level = 90
ygrid = 500
ax.scatter(levels, alt, marker='*')
ax.set_xticks(range(0, max_level + 1, 5))
ax.set_xticks(range(1, max_level), minor=True)
ax.set_yticks(range(0, max_alt + 1, 1000))
ax.set_yticks(range(0, max_alt, ygrid), minor=True)
ax.set_ylim(min_alt, max_alt)
ax.set_xlim(min_level, max_level)
ax.grid(axis='x', which='major')
ax.grid(axis='y', which='both')
ax.plot([0,0], [min_alt, alt[0]], color='k')
ax.plot([min_level, max_level], [0,0], color='k')
```
